# IRIS DATASET


---


In the following code cell, let us read the Iris Dataset. First, let us import the required Python libraries: Pandas and Matplotlib. Then, let us read the csv file and display it.

```python
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv("https://bit.ly/4nejNue")
iris
```

## Understanding the shape and structure of the dataset:


---


The following code cell tells us the number of rows and columns in the dataset:

```python
iris.shape
```

The following code cell displays the first 5 rows of the dataset:

```python
iris.head()
```

The following code cell counts the frequencies of each unique value of the species column:

```python
iris.species.value_counts()
```

## Summary statistics of the dataset:


---


The following code cell calculates the mean of each numeric column:

```python
iris.iloc[:,:-1].mean() # Method 1: Selecting the numeric type columns using iloc
```

The following code cell calculates the median of each numeric type column:

```python
iris.loc[:,['sepal_length','sepal_width','petal_length','petal_width']].median() # Method 2: Selecting the numeric type columns using loc
```

The following code cell calculates the standard deviation of each numeric type column:

```python
iris.std(numeric_only=True) # Method 3: Selecting the numeric type columns using numeric_only
```

The following code cell describes the basic summary statistics of the numeric type columns in our dataset:

```python
iris.describe()
```

The following code cell describes the summary statistics of the string type column i.e. the species column of our dataset:

```python
iris.species.describe()
```

## Plotting Histograms:


---


The following cell plots histograms for each numeric type column of the dataset:

```python
for column in iris.columns[:-1]:
  plt.figure()
  plt.hist(iris[column],bins=10,color='pink',edgecolor='black')
  plt.title(f"{column} Distribution")
  plt.xlabel(f"{column} (cm)")
  plt.ylabel('Frequency')
  plt.show()
```
